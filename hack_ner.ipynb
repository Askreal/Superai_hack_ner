```python
labels_lst20 = lst20_dataset['train'].features['ner_tags'].feature.names
labels_map_lst20 = {i: labels_lst20 for i, labels_lst20 in enumerate(labels_lst20)}
labels_lst20
```

```python
def transform_lst20_to_df(split):
    sentences = []
    for i, row in enumerate(split):
        words = row['tokens']
        labels = [labels_map_lst20[ner] for ner in row['ner_tags']]
        sentence_id = [i] * len(words)  # Keep sentence_id
        pos_tags = row['pos_tags']
        sentences.extend(zip(sentence_id, words, labels))  # Include sentence_id
    return pd.DataFrame(sentences, columns=["sentence_id", "words", "labels"])  # Include sentence_id column


def transform_superai_to_df(split):
    sentences = []
    sentence_id_counter = 0  # Counter for sentence_id
    for i, row in enumerate(split):
        # print(f"I : {i} Row : {row}")
        tokens = row.split()
        if len(tokens) >= 3:
            words = row.split()[0]
            labels = row.split()[2]
        else:
            # print(f"Skipping row {i}")
            sentence_id_counter += 1  # Increment sentence_id for the next sentence
            continue
        sentences.extend(zip([sentence_id_counter], [words], [labels]))  # Include sentence_id
    return pd.DataFrame(sentences, columns=["sentence_id", "words", "labels"])  # Include sentence_id column
```

```python
# Transform train and validation splits
train_lst_df = transform_lst20_to_df(lst20_dataset['train'])
eval_lst_df = transform_lst20_to_df(lst20_dataset['validation'])
```

```python
# Transform train and validation splits
train_superai_df = transform_superai_to_df(superai_dataset['train']['text'])
eval_superai_df = transform_superai_to_df(superai_dataset_eval['test']['text'])
```

```python
model_args = NERArgs()
model_args.train_batch_size = 64  # Reduce if GPU runs out of memory
model_args.num_train_epochs = 15  # Lower to avoid overfitting
model_args.overwrite_output_dir = True
model_args.reprocess_input_data = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_steps = 500  # Evaluate every 500 steps
model_args.learning_rate = 3e-5  # Common for transformers
model_args.early_stopping_patience = 3  # Stop early if no improvement
model_args.save_best_model = True  # Saves best checkpoint
```

```python
valid_labels = labels_lst20  # Replace with your list of labels
filtered_train_df = combined_train_df[combined_train_df['labels'].isin(valid_labels)]
filtered_eval_df = combined_eval_df[combined_eval_df['labels'].isin(valid_labels)]
```

```python
# Initialize the NER model
model = NERModel(
    model_type="roberta",
    # model_type="auto",
    model_name="lst-nectec/HoogBERTa-NER-lst20",
    labels=labels_lst20,
    args=model_args,
    use_cuda=True,
    ignore_mismatched_sizes=True
)
```

```python
model.train_model(filtered_train_df , eval_data=filtered_eval_df)
```

```python
result, model_outputs, preds_list = model.eval_model(filtered_eval_df)
result
```
